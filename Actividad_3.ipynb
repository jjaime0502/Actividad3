```python
data1=data.copy()
data1.info()
```

##Columna Ventas Precios Corrientes

```python
#Realizamos diagrama de caja o bigote
fig=plt.figure(figsize=(10, 6))
plt.boxplot(data["ventas_precios_corrientes"])
plt.title("Outliers de ventas_precios_corrientes ")
plt.show() #Dibujamos el diagrama
```

```python
#M√©todo aplicando desviaci√≥n estandar. Encuentro los valores extremos
y=data1["ventas_precios_corrientes"]
Limite_Superior=y.mean() +2.5*y.std()
Limite_Inferior=y.mean() -2.5*y.std()
print("L√≠mite superior permitido", Limite_Superior)
print("L√≠mite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Outliers
outliers=data1[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Ajustar m√°ximo de filas
pd.options.display.max_rows=None
```

```python
#Obtenemos datos limpios
data_clean_ventas_corrientes=data1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data_clean_ventas_corrientes
```

```python
#Realizamos diagrama de caja o bigote
fig=plt.figure(figsize=(5, 3))
plt.boxplot(data_clean_ventas_corrientes["ventas_precios_corrientes"])
plt.title("ventas_precios_corrientes")
plt.show() #Dibujamos el diagrama
```

## Ventas precios constantes

```python
#Realizamos diagrama de caja o bigote
fig=plt.figure(figsize=(10, 6))
plt.boxplot(data["ventas_precios_constantes"])
plt.title("Outliers de ventas_precios_constantes")
plt.show() #Dibujamos el diagrama
```

```python
from numpy.lib.function_base import percentile
#M√©todo aplicando Cuartiles. Encuentro cuartiles 0.25  0.75
y=data["ventas_precios_constantes"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 +1.5*iqr
Limite_Inferior_iqr=percentile25 -1.5*iqr
print("L√≠mite superior permitido", Limite_Superior_iqr)
print("L√≠mite inferior permitido", Limite_Inferior_iqr)
```

```python
outliers_iqr=data1[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_ventas_constantes=data1[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_ventas_constantes
```

```python
#Realizamos diagrama de caja o bigote
fig=plt.figure(figsize=(10, 6))
plt.boxplot(data_clean_ventas_constantes["ventas_precios_constantes"])
plt.title("Outliers de ventas_constantes")
plt.show() #Dibujamos el diagrama
```
